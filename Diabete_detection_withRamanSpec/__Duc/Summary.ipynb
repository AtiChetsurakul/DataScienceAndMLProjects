## Deep convolutional neural networks for Raman spectrum recognition: a unified solution

* Propose a unified solution to identify of chemical species in which a convolutional neural network is trained to automatically identify substances according to their Raman spectrum *without the need for preprocessing*

* Focusing on multivariate methods as well as *CNNs* which combines preprocessing, feature extraction and classification in a single architecture.

* Adopting a variant of LeNet, comprising pyramid-shaped convolutional layers for feature extraction and two fully-connected layers for classification.

* Training with one-dimensional convolutional kernels in the CNN.

* Using LeakyReLU nonlinearity which is followed by a max-pooling layer. After that, the upper layers of the CNN are fully connected implemented with tanh and followed by the softmax.

* To avoid overfitting, applying batch normalization after each layer and dropout after the first fully connected layer
